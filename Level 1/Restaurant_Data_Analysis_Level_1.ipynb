<h1>Distribution of Categorical Variables</h1>

```python
import seaborn as sns

ax = sns.countplot(x= df['Country Code'])
ax.set(xlabel='Country Codes', ylabel='Number of Restaurants')
ax.set_title('Distribution of Restaurants by Country Codes')
```

```python
ax = sns.countplot(y=df['City'], order=df.City.value_counts().iloc[:10].index)
ax.set(xlabel='Number of Restaurants', ylabel='Name of Cities')
ax.set_title('Top 10 Cities with Highest Number of Restaurants')
```

```python
ax = sns.countplot(y=df['Cuisines'], order=df.Cuisines.value_counts().iloc[:10].index)
ax.set(xlabel='Number of Restaurants', ylabel='Name of Cuisines')
ax.set_title('Top 10 Cities with Highest Number of Restaurants')
```

```python
top_cuisines = df['Cuisines'].value_counts().head(10)
print(f"Top 10 Cuisines with Highest Number of Restaurants: {top_cuisines}")

print(" ")

top_cities = df['City'].value_counts().head(10)
print(f"Top 10 Cities with Highest Number of Restaurants: {top_cities}")
```

<h1>Task 3: Geospatial Analysis</h1>
<ol><li>Visualize the locations of restaurants on a
map using latitude and longitude
information.</li>
<li>Analyze the distribution of restaurants across different cities or countries.</li>
<li>Determine if there is any correlation
between the restaurant's location and its
rating.</li>

</ol>

```python

from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df.Longitude, df.Latitude)
)

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))


gdf.plot(ax=world.plot("continent", legend = True, figsize=(14, 12)), marker='o', color='red', markersize=15)
```

```python
ax = sns.countplot(y=df['City'], order=df.City.value_counts().iloc[:10].index)
ax.set(xlabel='Number of Restaurants', ylabel='Name of Cuisines')
ax.set_title('Top 10 Cities with Highest Number of Restaurants')
```

```python
import matplotlib.pyplot as plt

correlation_matrix = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Restaurant's location and Rating")
plt.show()
```
